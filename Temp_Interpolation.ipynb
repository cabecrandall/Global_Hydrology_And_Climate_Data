```python
Temp_ts_path = 'Temp_TS/'
Temp_ts_exported = []

for file in os.listdir(Temp_ts_path):  
  file_path = os.path.join(Temp_ts_path,file)
  Temp_df = pd.read_csv(file_path)
  Temp_df = Temp_df.rename(columns={'Unnamed: 0':'Date'})
  Temp_df = Temp_df.rename(columns={'Region Daily Average LST':'Daily Average LST [C]'})

  Temp_df["Date"] = pd.to_datetime(Temp_df["Date"])
  Temp_df = Temp_df.set_index("Date")
  Temp_df = Temp_df.resample("D").asfreq()
  print(Temp_df)


  Temp_series = Temp_df['Daily Average LST [C]'].values.tolist()
  Temp_series = pd.Series(Temp_series)
  Temp_series = Temp_series.interpolate()
  Temp_list = Temp_series.to_list()
  Temp_df['Daily Average LST [C]'] = Temp_list

  file_PathName = 'Temp_TS_Filled/' + file
  Temp_df.to_csv(path_or_buf=file_PathName,index=True)
  break
```
